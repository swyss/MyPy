```python
# Define the known points
p1 = np.array([1, 5])
p2 = np.array([3, 4])
coefficients = np.polyfit(p1, p2, 1)
m = coefficients[0]
c = coefficients[1]
#
a = p1[1]-p2[1]
b = p2[0]-p1[0]
c = (p2[0]*p1[1])-(p1[0]*p2[1])
#
v_p = np.array([c/a, 0])  # np.array([0, c/b])
v_u = np.array([-b, a])
#
v_n = np.array([a, b])
v_p = np.array([c/a, 0])  # np.array([0, c/b])
#

print("HAUPTFORM: y = {m}x + {c}".format(m=m, c=c))
print("---")
print("PARAMETERFORM: [x1 x2]={v_p} + \u03BB * {v_u}".format(v_p=v_p, v_u=v_u))
print("---")
print("NORMALFORM: {v_n} * ( [x1 x2] - {v_p} ) = 0".format(v_n=v_n, v_p=v_p))
print("---")
print("KOORDINATENFORM: {a}x + {b}y = {c}".format(a=a, b=b, c=c))
print("---")
```

## Ebenengleichung

```python
p1 = np.array([1, 2, 3])
p2 = np.array([4, 6, 9])
p3 = np.array([12, 11, 9])

# These two vectors are in the plane
v1 = p3 - p1
v2 = p2 - p1

# the cross product is a vector normal to the plane
cp = np.cross(v1, v2)
a, b, c = cp

# This evaluates a * x3 + b * y3 + c * z3 which equals d
d = np.dot(cp, p3)
#
v_p = np.array([c/a, 0])  # np.array([0, c/b])
v_u = np.array([-b, a, 0])
v_v = np.array([0, -c, b])
#
v_n = np.array([a, b, c])
v_p = np.array([d/a, 0, 0])  # np.array([0, d/b, 0]) np.array([0, 0, d/c])
#

print(
    "PARAMETERFORM: [x1 x2 x3] = {v_p} + s * {v_u} + t * {v_v}".format(v_p=v_p, v_u=v_u, v_v=v_v))
print("---")
print(
    "NORMALFORM: ( [x1 x2 x3] - {v_p} ) * {v_n} = 0".format(v_n=v_n, v_p=v_p))
print("---")
print('KOORDINATENFORM:  {0}x + {1}y + {2}z = {3}'.format(a, b, c, d))
print("---")
```
